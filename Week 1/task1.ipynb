```python
import numpy as np
import matplotlib.pyplot as plt
import math 
```

```python
def f(x):
    return (math.e)**x
```

```python
def g(x):
    return 1 + x + 0.5*x**2 + 0.1667*x**3
```

```python
x_range = np.linspace(-0.5, 0.5, 1000)
```

```python
plt.plot(x_range, f(x_range), x_range, g(x_range))
plt.legend(['f(x)','g(x)'])
plt.show()
```
